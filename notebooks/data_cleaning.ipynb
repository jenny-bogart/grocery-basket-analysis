# Cleaning Code
Takes Grocery Store Dataset (https://www.kaggle.com/datasets/pratyushpuri/grocery-store-sales-dataset-in-2025-1900-record) and standardizes it.
Final Output:

1. customer_id: unique ID for customer: integer
2. store_name: grocery store name: string
3. transaction_date: date of purchase: date
4. aisle: aisle where product is located: string
5. product_name: name of product: string
6. quantity: number of product bought: float
7. unit_price: price per unit of product: float
8. total_amount: total price of product: float
9. discount_amount: discount on product: float
10. final_amount: final amount (after discount): float
11. loyalty_points: loyalty points of customer: integer

```python
# Read grocery store dataset to a Dataframe
!unzip grocery-store-sales-dataset-in-2025-1900-record.zip
import pandas as pd
df = pd.read_csv("grocery_chain_data.csv")
df.head()
```

```python
# Drop Dulicates
df.drop_duplicates(inplace=True)
```

```python
from datetime import date
df['customer_id'] = df['customer_id'].astype(int)
df['store_name'] = df['store_name'].astype(str)
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
df['aisle'] = df['aisle'].astype(str)
df['product_name'] = df['product_name'].astype(str)
df['quantity'] = df['quantity'].astype(float)
df['unit_price'] = df['unit_price'].astype(float)
df['total_amount'] = df['total_amount'].astype(float)
df['discount_amount'] = df['discount_amount'].astype(float)
df['final_amount'] = df['final_amount'].astype(float)
df['loyalty_points'] = df['loyalty_points'].astype(int)

```

```python
df.isnull().sum()
```

```python
# Save to CSV
df.to_csv("grocery_transactions_clean.csv", index=False)
```
